```python
from ultralytics import YOLO

# Load trained model
model = YOLO("runs/detect/train/weights/best.pt")  

# Run inference on a test image
image_path = "/kaggle/input/yolo-floor-2nd/Floor-plan-segmentation.v17i.yolov8/test/images/Screenshot-2023-12-06-215621_png.rf.15ed8c12e9685e683d684ce95f098398.jpg"
results = model(image_path)

# Get class names from the model
class_names = model.names

# Extract bounding box coordinates and class names
detections = []
for result in results:
    for box in result.boxes:
        x1, y1, x2, y2 = box.xyxy[0].tolist()  # Get bounding box coordinates
        cls_id = int(box.cls[0].item())  # Get class ID
        class_name = class_names[cls_id]  # Get class name
        detections.append({"class": class_name, "bbox": [x1, y1, x2, y2]})

# Print results
for det in detections:
    print(det)
```

```python
import cv2
import matplotlib.pyplot as plt
import random

# Load image
image = cv2.imread(image_path)
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)  # Convert BGR to RGB for correct display

# Assign a unique color for each class
class_colors = {}
for class_name in class_names.values():
    class_colors[class_name] = [random.randint(0, 255) for _ in range(3)]

# Draw bounding boxes with class-specific colors
for det in detections:
    x1, y1, x2, y2 = map(int, det["bbox"])
    class_name = det["class"]
    color = class_colors[class_name]  # Use the same color for the same class
    cv2.rectangle(image, (x1, y1), (x2, y2), color, 2)  # Draw bounding box

# Show the image
plt.figure(figsize=(10, 10))
plt.imshow(image)
plt.axis("off")
plt.show()
```

## In local machine

```python
import cv2
import torch
from ultralytics import YOLO

# Load YOLO Model
model = YOLO(r"E:\Floor-Plan-project\best.pt")  

# Load Image
image_path = r"E:\Floor-Plan-project\WhatsApp Image 2025-03-24 at 10.18.57 PM.jpeg"
image = cv2.imread(image_path)

# Preprocessing
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)  # Convert BGR to RGB 
image = cv2.resize(image, (640, 640))  # Resize to YOLO's input size
image = image / 255.0  # Normalize pixel values (0 to 1)
image = torch.tensor(image, dtype=torch.float32).permute(2, 0, 1).unsqueeze(0)  # Convert to Tensor

# Run YOLO Inference
results = model(image)

# Display Results
for result in results:
    result.show()
```
